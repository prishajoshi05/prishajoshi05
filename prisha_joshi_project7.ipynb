## Question 1

```r
# code here
library(data.table)
df <- fread("/anvil/projects/tdm/data/death_records/DeathRecords.csv")
a<- cut(df$Age , 
          breaks=c(-Inf,18,25, 35,55,150,Inf),
        right=TRUE,
        dig.lab = 4)
ab <- table(a)
ab
P<- cut(df$Age , 
        breaks=c(-Inf,18,25, 35,55,150,Inf),
        labels=c("youth","young adult","adult","middle age adult","senior adult","unknown"),
        right=TRUE,
       dig.lab=4)
with_labels <- table(P)
with_labels
barplot(with_labels, 
        main = " Age groups at the time of death",
        xlab = " Age groups",
        ylab = "Number of people")
          
                                                          
```

the records show that 36033 people died in their youth, 27691 people died when they were young adults , 49540 people died when they were adults , 271181 people died when they were senior adults and there were 571 deaths that were unknown. from the graph we can see an increase as the age group increases and then shoot up when the age group is senior adults. this data makes sense as naturally one would expect the most number of deaths to happen when you're old.
