## Question 1

```r
library(data.table)
myindyflights <- function(myyear) {
    myDF <- fread(paste0("/anvil/projects/tdm/data/flights/subset/", myyear, ".csv"))
    myvalue <- table(myDF$Origin)['IND']
    names(myvalue) <- myyear
    return(myvalue)
}
myresults <- sapply(1987:2008, myindyflights)
sum(myresults)
plot(names(myresults), myresults)
```

the plot shows us how the total number of flights starting from IND changed from 1987 to 2008. as we can see there is a sort of curve in the graph. in the year 1987 , there were less than 10000 flights that started and that increased drastically in the year 1988 when approx 3500 flights started from IND, and the number increases till the year 1193 and stoops down in the late 90's to go back up again until the sudden drop of 1500 flight happens in 2008. From this graph , there isn't any particar trend or relationship that can be inferred about the years and the number of flights. 

## Question 2

```r
library(data.table)
myflights <- function(myyear) {
    myDF <- fread(paste0("/anvil/projects/tdm/data/flights/subset/", myyear, ".csv"))
    myvalue <- table(myDF$Origin)
    return(myvalue)
}
myresults <- sapply(1987:2008, myflights)
v <- unlist(myresults)
tapply(v, names(v), sum)
dotchart(tail(sort(tapply(v, names(v), sum)), n=10))
```

we do notice a slight trend in this graph here where the airport in the tenth place has the smallest number of flights starting and ORD:the first on the list has the largest number of flights starting from that airport. however , the difference in the number of flights between the 3rd and the 4th airports changes drastically as compared to the rest. 

## Question 3

```r
library(data.table)
myindydonations <- function(myyear) {
    myDF <- fread(paste0("/anvil/projects/tdm/data/election/itcont", myyear, ".txt"), quote="", select = c(10,15))
    names(myDF) <- c("state", "donation")
    myvalue <- tapply(myDF$donation, myDF$state, sum)['IN']
    names(myvalue) <- myyear
    return(myvalue)
}
myresults <- sapply( seq(1980,2018,by=2), myindydonations )
plot(names(myresults), myresults)
```

the amount of money donated by the state of indiana per election also follows a trend , for most of the years. where the least amount was donated in the year 1980 and the most was in 2018. there are a few outliers in the trend in a few years.
