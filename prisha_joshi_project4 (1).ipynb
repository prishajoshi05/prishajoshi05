```r
# code here
library(data.table)
df2 <- fread("/anvil/projects/tdm/data/election/itcont1980.txt", quote="")
names(df2) <- c("CMTE_ID", "AMNDT_IND", "RPT_TP", "TRANSACTION_PGI", "IMAGE_NUM", "TRANSACTION_TP", "ENTITY_TP", "NAME", "CITY", "STATE", "ZIP_CODE", "EMPLOYER", "OCCUPATION", "TRANSACTION_DT", "TRANSACTION_AMT", "OTHER_ID", "TRAN_ID", "FILE_NUM", "MEMO_CD", "MEMO_TEXT", "SUB_ID")
merge <- paste(df2$CITY, df2$STATE, sep=",")
z<-tapply(df2$TRANSACTION_AMT , merge , sum)
w <- sort (z,decreasing = TRUE)
print(head(w, 20))
```

```r
# code here
library(data.table)
df3<- fread("/anvil/projects/tdm/data/flights/subset/1990.csv")
date <- paste(df3$Month, df3$DayofMonth , df3$Year ,sep="/")
d <-tapply(df3$DepDelay, date, mean, na.rm = TRUE)
p<-sort(d)
print(tail(p, 6))

```
